```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Data
age = np.array([3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17])
wing_length = np.array([1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5.0])
```

```python
# Task 1: Plot the relationship between Age and Wing Length
plt.scatter(age, wing_length)
plt.xlabel('Age')
plt.ylabel('Wing Length')
plt.title('Age vs. Wing Length')
plt.show()
```

```python
# Task 2: Calculate and plot the regression line
slope, intercept, r_value, p_value, std_err = linregress(age, wing_length)
regression_line = slope * age + intercept

plt.scatter(age, wing_length, label='Data')
plt.plot(age, regression_line, color='red', label='Regression Line')
plt.xlabel('Age')
plt.ylabel('Wing Length')
plt.title('Age vs. Wing Length with Regression Line')
plt.legend()
plt.show()
```

```python
# Task 3: Test if we can reject H0 : b = 0
# Null Hypothesis (H0): The slope is zero (no relationship)
# If p-value is small (typically < 0.05), we reject the null hypothesis
if p_value < 0.05:
    print(f"We reject H0. The slope is significantly different from zero (p-value={p_value}).")
else:
    print(f"We cannot reject H0. The slope is not significantly different from zero (p-value={p_value}).")
```

```python
# Task 4: Calculate and plot the confidence intervals on the slope of the regression
from scipy.stats import t

# Calculate standard error of the slope
n = len(age)
df = n - 2  # degrees of freedom
t_critical = t.ppf(0.975, df)  # 95% confidence interval, two-tailed
std_err_slope = std_err / np.sqrt(np.sum((age - np.mean(age))**2))

# Calculate confidence interval
conf_interval = (slope - t_critical * std_err_slope, slope + t_critical * std_err_slope)

# Plot the data, regression line, and confidence interval
plt.scatter(age, wing_length, label='Data')
plt.plot(age, regression_line, color='red', label='Regression Line')
plt.fill_between(age, slope*age + conf_interval[0], slope*age + conf_interval[1], color='gray', alpha=0.5, label='Confidence Interval (95%)')
plt.xlabel('Age')
plt.ylabel('Wing Length')
plt.title('Age vs. Wing Length with Regression Line and Confidence Interval')
plt.legend()
plt.show()

# Print confidence interval
print(f"Confidence Interval for the Slope (95%): ({conf_interval[0]}, {conf_interval[1]})")
```

```python
# Task 5: Calculate r^2 (coefficient of determination)
r_squared = r_value**2
print(f"The coefficient of determination (r^2) is: {r_squared}")
```

```python
# Task 6: Calculate Pearson's r
print(f"Pearson's r is: {r_value}")
```
